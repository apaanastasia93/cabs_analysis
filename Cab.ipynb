Let's compare number of active/loyal users each company has. Loyal user is a user who committed at least 10 trips with the company.

```python
cabs
```

```python
active_users = cabs[['Customer ID','Gender','Age','Income (USD/Month)']]
```

```python
active_users1 = cabs[['Customer ID','Trip_Count','Cost of Trip','Company']]
```

```python
active_users1=active_users1.groupby(by=['Customer ID','Company']).sum().reset_index()
active_users1=active_users1.merge(active_users, on='Customer ID')
active_users1['Customer_Count']=1
```

```python
active_users1 = active_users1.drop_duplicates()
```

```python
active_users = active_users1.loc[active_users1.Trip_Count>=10]
active_users.drop(columns=['Customer ID','Trip_Count','Cost of Trip','Gender','Age','Income (USD/Month)']).groupby(by='Company').sum()
```

**Yellow Cab** has 10 715 loyal users while **Pink Cab** has only 492.

#### What is a companies' target audience?

```python
plt.figure(figsize=(10,6))
sns.barplot(data=active_users1, hue='Gender', y='Trip_Count', x='Company', estimator=sum).set(title='Target Audience by Gender')
```

**Male** users get a taxi more often than **female** users.

```python
active_users1['Income (USD/Month)'].describe()
```

```python
active_users2 = []
for row in active_users1['Income (USD/Month)']:
    if row <=8299:    active_users2.append('Low Class')
    elif row>8299 and row<=15000:   active_users2.append('Middle Class')
    else:           active_users2.append('Upper Middle Class')
```

```python
active_users1['Level_Of_Income'] = active_users2
active_users1
```

```python
active_users1.loc[active_users1['Income (USD/Month)']>=21029]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=active_users1, hue='Level_Of_Income', y='Trip_Count', x='Company', estimator=sum).set(title='Target Audience by Income')
```

**Wealthy Class** users use a taxi more often than other categories.

```python
active_users3 = []
for row in active_users1['Age']:
    if row <=30:    active_users3.append('20-30 years old')
    elif row>30 and row<=40:   active_users3.append('31-40 years old')
    elif row>40 and row<=50:   active_users3.append('41-50 years old')
    else:           active_users3.append('51+ years old')
```

```python
active_users1['Age_category'] = active_users3
active_users1
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=active_users1, hue='Age_category', y='Trip_Count', x='Company', estimator=sum).set(title='Target Audience by Age')
```

**20-30 years old** users and **31-40 years old** users get a taxi more often than other categories.

Therefore, the target audience: **Young male (20-40) years old with an income 15000 dollars per month or higher**.
